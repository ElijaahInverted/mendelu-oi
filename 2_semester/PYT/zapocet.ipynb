2. Napiš kód pro funkci suda_licha s variabilním množstvím parametrů, která vrátí dva seznamy (sudá, lichá), a dekorátor ponechávající jen přirozená čísla.

```python
def jen_prirozena(func):
    def wrapper(*args):
        filtrovane = [x for x in args if isinstance(x, int) and x > 0]
        return func(*filtrovane)
    return wrapper

@jen_prirozena
def suda_licha(*args):
    suda = [x for x in args if x % 2 == 0]
    licha = [x for x in args if x % 2 != 0]
    return suda, licha
```

```python
# BEZ dekorátoru by suda_licha dostala všechny argumenty:
# suda_licha(3, -1, 'ahoj', 4.5, 2, 0, 7)  ->  args = (3, -1, 'ahoj', 4.5, 2, 0, 7)

# S dekorátorem @jen_prirozena wrapper nejdřív odfiltruje:
#   isinstance(x, int) and x > 0  =>  zůstanou jen [3, 2, 7]
# a teprve ty předá do suda_licha

result = suda_licha(3, -1, 'ahoj', 4.5, 2, 0, 7)
print('suda:', result[0])   # [2]
print('licha:', result[1])  # [3, 7]
```

3. Napiš kód pro třídu ZvireVDomacnosti s vlastnostmi druh, jméno, id, seznam_ockovani, metodou ockovani, metodou pro výpis a magickou metodou pro chování v kontextu řetězce.

```python
class ZvireVDomacnosti:
    def __init__(self, druh, jmeno, id):
        self.druh = druh
        self.jmeno = jmeno
        self.id = id
        self.seznam_ockovani = []

    def ockovani(self, typ, datum):
        self.seznam.ockovani.append((typ, datum))

    def vypis(self):
        print(f"(self.jmeno), ({self.druh}), ID: {self.id}")
```
